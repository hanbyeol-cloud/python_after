```python
import requests # 웹 브라우저 처럼 = > text(tag+javascript)
import pandas as pd
from bs4 import BeautifulSoup # tag파싱 => DOM (계층적으로 노드를 구성) : select , find
import re # regular expression 정규표현식 : 패턴으로 인식
```

```python
class HTMLParser: #테이블을 동시에 파싱
    def parse_url(self, url):
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        return soup
```

```python
url = "http://media.daum.net/ranking/popular/"
print(url)
hp = HTMLParser()
soup=hp.parse_url(url)
```

```python
content=soup.find_all('div', {"class" : "cont_thumb"})
content[0].find_all('span', {"class" : "info_news"})
```

```python
r=len(content)
for i in range(r):
    print(content.find_all('span')[i])
    #body.append(i.find_all('span')[0].get_text())
```

```python
head=[]
body=[]
content[0].find_all('span', {"class" : "info_news"}).get_text()
for i in range(len(content)):
    head.append(content[i].find_all('span', {"class" : "info_news"}))
    body.append(content[i].find_all('span', {"class" : "link_txt"}))

articles = {'기사제목' : head,
            '기사내용' : body}
articles
```
